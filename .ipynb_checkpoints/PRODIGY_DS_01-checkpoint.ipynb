```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('PopulationTrend.csv',na_values=['??','????'], header=None)
print(data.head())
```

```python
# As we can see first four rows contain nan values they are no use to us. So we will just drop them
population = pd.read_csv('PopulationTrend.csv',na_values=['??','????'],skiprows=4)
population
```

```python
# Now we have correct dataset for data visualisation.
print(population.describe())
```

```python
print(population.dtypes)
```

```python
# lets draw bar graph for population of india from year 1960

# Filter the data for Afghanistan
india_data = population[population['Country Name'] == 'India']
# Extract the years and population values, ignoring non-numeric columns
years = india_data.columns[4:]  # Assuming the first 4 columns are non-numeric
population_of_India = india_data.iloc[0, 4:].astype(float)  # Convert the population data to float
# Plot the bar graph
plt.figure(figsize=(12, 6))
sns.barplot(x=years, y=population_of_India, palette='viridis',hue=years)
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population of India Over the Years')
plt.xticks(rotation=90)  # Rotate the x-axis labels for better readability
plt.show()
```

```python
# Plotting population of different neighbouring countries of india in from year 2022
```

```python
#List of contries for the graph.
countries_list = ['Pakistan', 'India', 'China', 'Nepal', 'Sri Lanka', 'Maldives','Myanmar']

# Seperate the data of desired countries
seperated_data = population[population['Country Name'].isin(countries_list)]
seperated_data
```

```python
plt.bar (seperated_data['Country Name'], seperated_data['2022'])
plt.xlabel ('Country Name')
plt.ylabel ('Population in 2022')
plt.title('Population of selected countries in 2022')
plt.xticks(rotation=45, ha='right') #Rotating the x-axis labels for readability
plt.show()
```

```python
# lets have a look to Indian population trend over the years. For that we will draw a scaterplot 
```

```python
# First seperate data of India
indian_population = population[population['Country Name'] == 'India']
indian_population
```

```python
# Extract years and population
years = indian_population.columns[4:-1] # Excluding the last column which is '2023' with NaN values
population = indian_population.iloc[0,4:-1].astype(float) # Convert the population data to float

sns.scatterplot(x=years, y=population, palette='viridis')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population of India Over the Years')
plt.xticks(rotation=90)  # Rotate the x-axis labels for better readability
plt.show()
```

```python
# From this scaterplot we can say that Indian population is continuously increasing
```
